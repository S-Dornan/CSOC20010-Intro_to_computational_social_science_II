```python
#1
import numpy as np
from scipy.signal import savgol_filter

from matplotlib import pyplot as plt 

# You miight need to install pageviewapi 
# You can do so by using the following command in the terminal, or search how to install python libraries 
# in windows/macOS depending on your operating system

#   pip install git+https://github.com/Commonists/pageview-api.git

import pageviewapi

```

```python
#2
# Here you can change the parameters of your querry
#date format: YYYYMMDD

j=pageviewapi.per_article('en.wikipedia', 'Taha_Yasseri', '20181001', '20220201',
                        access='all-access', agent='all-agents', granularity='daily')
```

```python
#3 
# The outcome of the pageviewapi is a JSON file
# here we read its items one by one and copy the "view" values in a new variable (v)

items=j['items']
v=list()
for item in items:
    v.append(item['views'])
```

```python
#print(v)
```

```python
#4
# we use the code from previus homework to plot the view statsitics
# please remember to change the lables, legends and the title accordingly

plt.title("SI model") 
plt.xlabel("Simmulation Steps") 
plt.ylabel("Ratio of Infected Nodes") 
plt.plot(v)
plt.legend(["hs","config","er","ba","ws"])
plt.savefig('SI.png',dpi=300)
plt.show()
```

```python
#5
# Bonus! here you can smoothen the curves using the Savitzky-Golay filter

v_smooth = savgol_filter(v, 21, 5)
```

```python
plt.plot(v_smooth)
plt.show()
```
